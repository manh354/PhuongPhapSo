#### Nội suy Lagrange

Đánh giá sai số của Lagrange <br>
<br>
<br>
 <br>
- Độ phức tạp thuật toán là $O(n^{3})$ => tốc độ xử lý của phép toán chậm <br>
- Sai số của **đa thức nội suy phụ thuộc vào các mốc nội suy** => **không thể khống chế được sai số** => chuyển đến sử dụng **NS ngược** <br>
<br>
**Ưu:** <br> 
- Có thể sử dụng để tìm giá trị của hàm ngay cả khi các mốc nội suy của hàm không cách đều <br>
- Dễ dàng tính toán và thiết kế, tìm biến x độc lập, của gtri hàm số cho trước <br>
**Nhược:** <br>
- Bổ sung thêm mốc ns => phải tính tonas lại từ đầu (đổi bậc của đa thức làm thay đổi tính toán cho toàn hạng tử) <br>
- Đối với đa thức bậc cao thì tính chậm $O(n^{3})$ <br>


##### Nội suy Newton

Độ phức tạp **$O(n^{2})$** (nhanh hơn Lagrange **$O(n^{3})$** )

Largrange dùng được cho hàm đơn điệu (monotomic function), Newton 0 sử dụng cho đơn điệu (mất tính lặp của phương pháp) <br>


 <br> 
(hình ảnh cho thấy biên bên phải xấp xỉ giá trị tốt) <br>
=> phương pháp nội suy newton thường **xấp xỉ tốt ở biên** 


**Ưu:** <br> 
- Bổ sung mốc nội suy không phải tính toán lại từ đầu <br>
- Khối lượng tính toán thấp hơn Lagrange<br>
- Xấp xỉ tốt với đa thức bậc cao <br>

**Nhược:** <br>
- Dễ gây ra hiện tượng tràn dữ liệu <br>
- Dữ liệu $x_{0},..,x_{n}$ phải **cách đều** <br>
- Xấp xỉ tốt ở khoảng nội suy => khoảng ngoại suy sai lệch rất lớn <br>
